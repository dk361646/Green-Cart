```python
df_clean = df[['title', 'description', 'category', 'material', 'brand', 'rating']]
```

```python
import re
import string
```

```python
def clean_text(text):
    if isinstance(text, str):
        text = text.lower()  # lowercase
        text = re.sub(r'\d+', '', text)  # remove numbers
        text = text.translate(str.maketrans('', '', string.punctuation))
        text = text.strip()  # remove leading/trailing spaces
        text = re.sub('\s+', ' ', text)  # remove extra spaces
        return text
    else:
        return ""
df_clean.loc[:, 'title'] = df_clean['title'].apply(clean_text)
df_clean.loc[:, 'description'] = df_clean['description'].apply(clean_text)
df_clean.loc[:, 'material'] = df_clean['material'].apply(clean_text)
df_clean.loc[:, 'brand'] = df_clean['brand'].apply(clean_text)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
#combining title and description to get richer data
df_clean.loc[:,'text'] = df_clean.loc[:, 'title'] + " " + df_clean.loc[:,'description']
```

```python
#tokenizing to convert into numbers, oov = out of vocabulary
tokenizer = Tokenizer(num_words=10000, oov_token='<OOV>')
tokenizer.fit_on_texts(df_clean['text'])
```

```python
#sequencing and padding (adding 0s after (post))
sequences = tokenizer.texts_to_sequences(df_clean.loc[:,'text'])
padded_sequences = pad_sequences(sequences, padding='post', maxlen=100)
```

```python
# Define input layer (sequence of tokens)
input_seq = Input(shape=(100,))  # We are using maxlen=100

# Embedding layer to convert words to dense vectors
embedding_layer = Embedding(input_dim=10000, output_dim=128)(input_seq)

# LSTM layer to capture sequential relationships
lstm_layer = LSTM(128, return_sequences=False)(embedding_layer)

# Optional: Pooling layer to average out features
pooling_layer = GlobalAveragePooling1D()(embedding_layer)

# Fully connected dense layer to get final embedding representation
dense_layer = Dense(128, activation='relu')(lstm_layer)

# Output layer (this can vary depending on the task)
output = Dense(1, activation='sigmoid')(dense_layer)

# Create the model
model = Model(inputs=input_seq, outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Fit the model to the padded sequences
model.fit(padded_sequences, np.ones(len(df_clean)), epochs=3, batch_size=32, validation_split=0.2)
```

```python
model.summary()
```

```python
# Create an embedding model that outputs from the Dense(128) layer
embedding_model = Model(inputs=model.input, outputs=model.get_layer(index=-2).output)

# Generate embeddings for all products
product_embeddings = embedding_model.predict(padded_sequences)

# Check the shape of embeddings
print(product_embeddings.shape)  # Should be (number of products, 128)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recommend_similar_products(product_index, embeddings, top_n=5):
    """
    Recommend top_n similar products based on cosine similarity.

    Args:
    - product_index: Index of the product for which recommendations are needed.
    - embeddings: Numpy array of product embeddings.
    - top_n: Number of recommendations to return.

    Returns:
    - DataFrame of recommended product titles, brands, materials.
    """
    # Step 1: Get the embedding of the selected product
    selected_embedding = embeddings[product_index].reshape(1, -1)

    # Step 2: Compute cosine similarity with all other product embeddings
    similarities = cosine_similarity(selected_embedding, embeddings)[0]

    # Step 3: Get indices of the most similar products
    similar_indices = np.argsort(similarities)[::-1][1:top_n+1]  # [1:] to skip itself

    # Step 4: Retrieve product information
    recommended_products = df_clean.iloc[similar_indices][['title', 'brand', 'material']]

    return recommended_products
```

```python
# Example: Recommend similar products
recommendations = recommend_similar_products(5, product_embeddings, top_n=5)

# Display the recommendations
print(recommendations)
```

```python
def get_sustainability_score(material_text):
    if isinstance(material_text, str):
        material_text = material_text.lower()

        if 'organic' in material_text or 'bamboo' in material_text or 'hemp' in material_text:
            return 1.0
        elif 'recycled' in material_text or 'sustainable' in material_text:
            return 0.8
        elif 'cotton' in material_text or 'linen' in material_text:
            return 0.7
        elif 'polyester' in material_text or 'nylon' in material_text or 'leather' in material_text:
            return 0.3
        else:
            return 0.5  # Neutral if unknown
    else:
        return 0.5  # Neutral for missing material
```

```python
# Add a new column for sustainability
df_clean['sustainability_score'] = df_clean['material'].apply(get_sustainability_score)

# Check the first few rows
df_clean[['material', 'sustainability_score']].head()
```

```python
sustainability_criteria = {
    "organic cotton": "Made from organic cotton, which uses fewer pesticides and less water, making it more environmentally friendly.",
    "recycled cotton": "Recycled cotton reduces textile waste and saves water and energy during production.",
    "bamboo": "Bamboo grows rapidly and requires no pesticides, making it a renewable and sustainable resource.",
    "hemp": "Hemp requires very little water and no pesticides. It's biodegradable and highly durable.",
    "linen": "Linen is made from flax, which grows with minimal water and no pesticides, making it eco-friendly.",
    "recycled polyester": "Recycled polyester is made from post-consumer plastic waste, reducing landfill use and carbon footprint.",
    "tencel": "Tencel is produced from sustainably sourced wood pulp in a closed-loop system, making it an eco-conscious fabric.",
    "modal": "Modal is a semi-synthetic fabric derived from beech tree pulp using a closed-loop production process.",
    "cork": "Cork is harvested from trees without cutting them down, making it a renewable, biodegradable material.",
    "jute": "Jute is a natural fiber that requires little fertilizer or pesticides, and it’s biodegradable.",
    "recycled materials": "This product uses recycled materials, helping reduce landfill waste and the need for virgin resources.",
    "biodegradable": "This product is made from biodegradable materials, allowing it to break down naturally and reduce waste.",
    "fair trade": "Certified fair trade, ensuring workers receive fair wages and work in ethical conditions.",
    "vegan": "This product contains no animal products or by-products, reducing harm to animals and the environment.",
    "natural rubber": "Natural rubber is biodegradable and sourced from trees, making it a more sustainable alternative to synthetic rubber.",
    "plant-based": "Made from plant-derived ingredients, reducing dependency on synthetic or fossil-based materials.",
    "compostable": "Designed to break down completely and safely in composting systems, reducing environmental impact.",
    "polyester": "Polyester is derived from petroleum and is non-biodegradable, contributing to microplastic pollution.",
    "nylon": "Nylon is energy-intensive to produce and not biodegradable, posing environmental concerns.",
    "acrylic": "Acrylic is made from fossil fuels and can release harmful fibers during washing.",
    "leather": "Traditional leather production involves toxic chemicals and high resource consumption.",
    "PVC": "Polyvinyl chloride is a synthetic plastic that is not biodegradable and can release harmful chemicals.",
    "synthetic rubber": "Synthetic rubber is derived from petroleum and contributes to environmental pollution.",
    "fur": "Fur involves animal cruelty and energy-intensive processing.",
    "down feathers": "Down can be unsustainable due to concerns about animal welfare unless responsibly sourced."
}

def sustainability_explanation(product_material):
    for material, explanation in sustainability_criteria.items():
        if material.lower() in product_material.lower():
            return explanation
    return "Sustainability details not found for this product."
```

```python
def get_sustainability_info(product):
    material = product['material']  # or description or brand
    explanation = sustainability_explanation(material)

    return {
        'explanation': explanation
    }
```

```python
def recommend_sustainable_products(product_index, embeddings, top_n=5, sustainability_weight=0.5, threshold=0.7):
    """
    Recommend products with sustainability awareness and green badges.
    """

    # Step 1: Check sustainability of selected product
    selected_product = df_clean.iloc[product_index]
    selected_sustainability = selected_product['sustainability_score']
    explaination = sustainability_explanation(selected_product['material'])

    if selected_sustainability >= threshold:
        print(f"✅ The selected product ('{selected_product['title']}') is sustainable! 🌱 (Score: {selected_sustainability}) (Explaination: {explaination})")
    else:
        print(f"⚠️ The selected product ('{selected_product['title']}') is NOT very sustainable. (Score: {selected_sustainability}) (Explaination: {explaination})")

    print("\n♻️ Recommending more sustainable similar products...\n")

    # Step 2: Get selected product embedding
    selected_embedding = embeddings[product_index].reshape(1, -1)

    # Step 3: Compute cosine similarity
    similarities = cosine_similarity(selected_embedding, embeddings)[0]

    # Step 4: Get sustainability scores
    sustainability_scores = df_clean['sustainability_score'].values

    # Step 5: Combine similarity and sustainability
    combined_scores = (1 - sustainability_weight) * similarities + sustainability_weight * sustainability_scores

    # Step 6: Sort and get top N
    recommended_indices = np.argsort(combined_scores)[::-1][1:top_n+1]

    # Step 7: Prepare recommendation list
    recommended_products = []

    for idx in recommended_indices:
        title = df_clean.iloc[idx]['title']
        brand = df_clean.iloc[idx]['brand']
        material = df_clean.iloc[idx]['material']
        score = df_clean.iloc[idx]['sustainability_score']

        # Add 🌱 badge if sustainable
        badge = "🌱" if score >= threshold else ""

        recommended_products.append({
            'Title': f"{title} {badge}",
            'Brand': brand,
            'Material': material,
            'Sustainability Score': round(score, 2)
        })

    # Step 8: Return as dataframe
    return pd.DataFrame(recommended_products)


```
